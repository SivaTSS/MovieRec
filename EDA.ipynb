```python
import pandas as pd
from surprise import Dataset, Reader
from surprise.model_selection import train_test_split
from surprise import SVD
from surprise import accuracy

# Assuming df is your DataFrame with columns 'userId', 'movieId', 'rating'
reader = Reader(rating_scale=(0.5, 5.0))
data = Dataset.load_from_df(df[['userId', 'movieId', 'rating']], reader)

# Split the dataset into training and testing sets
trainset, testset = train_test_split(data, test_size=0.2, random_state=42)

# Build an SVD model
model = SVD()

# Train the model
model.fit(trainset)

# Make predictions on the test set
predictions = model.test(testset)

# Evaluate the model
accuracy.rmse(predictions)
```

```python
# Recommend movies for a specific user (e.g., userId = 1)
user_id = 1
user_movies = df[df['userId'] == user_id]['movieId'].unique()
user_unseen_movies = df[~df['movieId'].isin(user_movies)]['movieId'].unique()

user_recommendations = []

for movie_id in user_unseen_movies:
    predicted_rating = model.predict(user_id, movie_id).est
    user_recommendations.append((movie_id, predicted_rating))

# Sort the recommendations by predicted rating
user_recommendations = sorted(user_recommendations, key=lambda x: x[1], reverse=True)

# Display the top N recommendations
top_n = 10
top_recommendations = user_recommendations[:top_n]

print(f"Top {top_n} recommendations for user {user_id}:")
for movie_id, predicted_rating in top_recommendations:
    movie_title = df[df['movieId'] == movie_id]['title'].values[0]
    print(f"Movie: {movie_title}, Predicted Rating: {predicted_rating}")
```

## Content-Based Filtering

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel


# Assuming df_movies is your DataFrame with columns 'movieId', 'title', 'genres'
# Combine relevant features into a single text column for each movie
df_movies['features'] = df_movies['title'] + ' ' + df_movies['genres'].fillna('')
df_movies_s = df_movies.sample(frac=0.1, random_state=42)
# Create a TF-IDF vectorizer to convert text data into numerical vectors
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(df_movies_s['features'])

# Calculate the cosine similarity between movies based on their features
cosine_similarity = linear_kernel(tfidf_matrix, tfidf_matrix)

def get_content_based_recommendations(movie_title, cosine_sim, df_movies_s):
    # Check if the movie title exists in the sampled DataFrame
    if movie_title not in df_movies_s['title'].values:
        return f"No information available for {movie_title}"

    # Get the index of the movie with the given title
    movie_index = df_movies_s[df_movies_s['title'] == movie_title].index[0]

    # Get the pairwise similarity scores with other movies
    sim_scores = list(enumerate(cosine_sim[movie_index]))

    # Sort the movies based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the movie indices of the top N similar movies
    top_indices = [i[0] for i in sim_scores[1:11]]

    # Return the titles of the top N similar movies
    return df_movies_s['title'].iloc[top_indices]

# Example: Get content-based recommendations for a movie
movie_title = "Jimmy Neutron: Boy Genius (2001)"
content_based_recommendations = get_content_based_recommendations(movie_title, cosine_similarity, df_movies_s)

print(f"Content-based recommendations for {movie_title}:")
print(content_based_recommendations)

```

## Hybrid Recommendation Systems

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
from surprise import Dataset, Reader
from surprise.model_selection import train_test_split
from surprise import KNNBasic
from surprise import SVD
from surprise import accuracy

# Load movies dataset
df_movies = pd.read_csv('path_to_movies_dataset.csv')
# Load ratings dataset
df_ratings = pd.read_csv('path_to_ratings_dataset.csv')

# Merge datasets on 'movieId'
df = pd.merge(df_ratings, df_movies, on='movieId')

# Create a TF-IDF vectorizer for content-based filtering
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(df_movies['title'] + ' ' + df_movies['genres'].fillna(''))

# Train collaborative filtering model
reader = Reader(rating_scale=(0.5, 5.0))
data = Dataset.load_from_df(df[['userId', 'movieId', 'rating']], reader)
trainset, testset = train_test_split(data, test_size=0.2, random_state=42)

# Build collaborative filtering model
collab_model = KNNBasic(sim_options={'name': 'cosine', 'user_based': True})
collab_model.fit(trainset)

# Build content-based filtering model
svd_model = SVD()
svd_model.fit(trainset)

# Function to get hybrid recommendations
def get_hybrid_recommendations(user_id, movie_title, collab_model, svd_model, df_movies, tfidf_vectorizer, tfidf_matrix):
    # Get collaborative filtering recommendations
    collab_recommendations = [i[0] for i in collab_model.get_neighbors(svd_model.predict(user_id, df_movies[df_movies['title'] == movie_title]['movieId'].values[0]).iid, k=10)]

    # Get content-based filtering recommendations
    movie_index = df_movies[df_movies['title'] == movie_title].index[0]
    content_based_scores = linear_kernel(tfidf_matrix[movie_index], tfidf_matrix).flatten()
    content_based_recommendations = [i for i in content_based_scores.argsort()[-11:-1][::-1]]

    # Combine recommendations from both models
    hybrid_recommendations = list(set(collab_recommendations + content_based_recommendations))

    # Return the titles of the top N hybrid recommendations
    return df_movies['title'].iloc[hybrid_recommendations]

# Example: Get hybrid recommendations for a user and a movie
user_id = 1
movie_title = 'The Dark Knight'
hybrid_recommendations = get_hybrid_recommendations(user_id, movie_title, collab_model, svd_model, df_movies, tfidf_vectorizer, tfidf_matrix)

print(f"Hybrid recommendations for user {user_id} and movie {movie_title}:")
print(hybrid_recommendations)
```
